```python
"""
Ejercicio 4:

Leer un número entero positivo y determinar la suma de sus dígitos pares. Por
ejemplo, en el número 124, los dígitos pares son el 2 y el 4 y su suma vale 6.
"""
#%%
"""
Ingresamos el número entero positivo y con el comando 'while', hacemos una 
iteración que deje analizar cada dígito de acuerdo al residuo que arroja. Si 
el dígito es divisible exactamente entre 2, es par y lo suma con los otros
posibles dígitos pares del número ingresado.
"""
n=int(input('Ingrese un número entero positivo: '))

p=0

while n/10 > 0:
    u = n % 10
    
    if u%2==0:
        p=p+u
        
    n = int(n/10)   
    
print('La suma de los dígitos pares es: ',p)   
```

```python
"""
Ejercicio 5:

Leer un número entero y determine el número de cincos que están consecutivos en
un número. Por ejemplo, en 5512355551 hay 6 cincos consecutivos.
"""
#%%
"""
Definimos una función, la cual nos ayudará a contar la cantidad de '5' que se 
encuentran consecutivos dentro del número ingresado. Con el comando 'while'
iteramos para desglosar el número y que se pueda leer dígito por dígito para 
comparar si es un '5' y si está seguido de otro '5'.
"""
def cantidad_cincos(x):
    s=0
    while x>0:
        d_1=x%10
        x=x//10
        d_2=x%10
        if d_1==5 and d_2==5:
            s+=2
            u=x//10
            d_3=u%10
            if d_2==5 and d_3==5:
                s+=1
                if d_1==5 and d_3==5:
                    s-=2
    return s
while True:
    try:
        n=int(input('Ingrese un número: '))
        if n<0:
            n*=(-1)
        break
    except ValueError:
        print('Inténtelo de nuevo, ingrese un número: ')

s=cantidad_cincos(n)
if s>0:
    print('El número de cincos consecutivos que contiene el número es: ',s)
else:
    print('El número no contiene cincos consecutivos.')
```

```python
"""
Ejercicio 6:

Leer un número entero y determinar si la suma de sus dígitos es un número de 
Fibonacci.
"""
#%%
"""
Defino una función que me ayude a contar los número desde el '1' que es donde 
comienza a formarse la suma para la cadena de Fibonacci y, con ayuda del 
comando 'while' iteramos de manera que se pueda llegar al número ingresado con
la cadena o el orden que se debe llevar en un número de Fibonacci, de manera 
que se sume el número tantas veces sea necesario para cumplir con el orden.
"""

def fib(x):
    n_fibonacci=[]
    a,b=1,1
    
    while a<=x:
        n_fibonacci.append(a)
        a,b=b,a+b
        
    return n_fibonacci
    
n=int(input('Ingrese un número entero: '))

suma=0

while n/10>0:
    u=n%10
    n=int(n/10)
    suma=suma+u

if suma in fib(suma):
    print('La suma de los dígitos es un número de Fibonacci. ')
    
else:
    print('La suma de los dígitos no es un número de Fibonacci. ')
```

```python
"""
Ejercicio 7:

Leer un número entero positivo y determine si existen en el número dígitos
repetidos. Reporte cuáles son los dígitos repetidos.
"""
#%%
"""
Con el comando'for' defino qué dígitos se estan repitiendo y se agregan a una
lista, la cual imprime el valor del dígito que se repite.
"""

n=input('Ingrese un número entero positivo: ')

u=[]
repetido=[]

for x in n:
    
	if x not in u:
		u.append(x)
        
	else:
        
		if x not in repetido:
			repetido.append(x)
            
print('Los dígitos que se repiten son: ',repetido)
```

```python
"""
Ejercicio 8:

Iterar a través de los primeros cien enteros positivos, buscando los múltiplos
de 3 e imprimiendo y almancenando en una lista hasta encontrar los primeros 15
de ellos. Una vez encontrados, continuar iterando en busca de los multiplos de
4 y almacenarlos en otra lista.
"""

#%%
"""
Se define el rango de los primeros cien enteros positivos. Luego de ello, se 
define hasta dónde va a contar los múltiplos del 3 y del 4. Defino una función 
para ejecutar la operación y definir si es o no múltiplo del 3 y posterior a 
este si es o no multiplo del 4. Con el comando 'for' determino y almaceno los
múltiplos del 3 que están dentro del rango definido anteriormente. Luego, se
hace lo mismo con los números restantes que son múltiplos del 4.
"""

r_i=1
r_f=101

lim_3=16
i_4=lim_3-5

mult_3= [3]
mult_4=[]

def multiplos(valor,mult):
    
    residuo=valor%mult
    
    if residuo==0:
        return True
    
    else:
        return False
    
for a in range (r_i,r_f):
    
    if multiplos(a,3):
        mult_3.append(a)
        
print('Los quince primeros multiplos del 3 son: ',mult_3[r_i:lim_3])

for b in range (r_i,r_f):
    
    if multiplos(b,4):
        mult_4.append(b)

print('Los multiplos restantes del 4, dentro del rango, son: ',mult_4[i_4:r_f])
```

```python
"""
Ejercicio 9:

Hacer un programa que lea las coordenadas (x1,y1,r1) y (x2,y2,r2) que
corresponden al centro y radio de dos círculos. Lea un punto de coordenadas 
(a,b). Determina si (a,b) está contenido:
a. Dentro del círculo 1
b. Dentro del círculo 2
c. Dentro de ambos círculos
d. Fuera de ambos círculos
Recordemos que el círculo es el lugar geométrico de los puntos del plano cuya
distancia a otro punto fijo, llamado centro, es menor o igual que una cantidad
constante llamada radio.
"""
#%%
"""
Definimos una función la cual calcula la distancia entre dos puntos. Con el 
comando 'while' abrimos una iteración en la cual se ingresarán las coordenadas
'x' y 'y' del centro del círculo 1. Luego de ello ingresamos el valor del radio
, este valor debe ser positivo. Hacemos el mismo proceso para el círculo 2.
Seguimos con ingresar las coordenadas que queremos saber si están o no en 
alguno de los círculos.
Finalmente, comprobamos si la distancia desde el centro del cículo hasta el
punto ingresado es igual o menos al radio, de ser así, éste punto ingresado
está dentro del círculo, de lo contrario, no lo está.
"""
def distancia(x,y,a,b):

    dist=(((x-a)**2)+((y-b)**2))**(1/2)
    return dist
  
while True :
    coordx_1= float(input('Ingrese la coordenada en X del círculo 1: '))
    break

while True :
    coordy_1= float(input('Ingrese la coordenada en Y del círculo 1: '))
    break
            
while True :
        r_1= float(input('Ingrese el radio del círculo 1: '))
        if r_1 >= 0 :
            break
    
while True :
    coordx_2= float(input('Ingresar la coordenada en X del círculo 2: '))
    break

while True :
    coordy_2= float(input('Ingresar la coordenada en Y del círculo 2: '))
    break
        
while True :
    r_2= float(input('Por favor ingrese el radio del círculo 2: '))
    if r_2 >= 0 :
        break

while True :
    a= float(input('Por favor ingrese la coordenada en X del punto: '))
    break

while True :
    b= float(input('Por favor ingrese la coordenada en Y del punto: '))
    break
        
distanciac_1= distancia(coordx_1, coordy_1, a, b)
distanciac_2= distancia(coordx_2, coordy_2, a, b)

if distanciac_1 <= r_1 and distanciac_2 <= r_2:
    print ('La coordenada', a,',', b, 'esta contenida en ambos círculos. ')
    
elif distanciac_1 > r_1 and distanciac_2 <= r_2:
    print ('La coordenada', a,',', b, 'esta contenida dentro del círculo 2. ')
    
elif distanciac_1 <= r_1 and distanciac_2 > r_2:
    print ('La coordenada', a,',', b, 'esta contenida dentro del círculo 1. ')
    
elif distanciac_1 > r_1 and distanciac_2 > r_2:
    print ('La coordenada', a,',', b, 'no esta contenido en ningún círculo. ')
```

```python
"""
Ejercicio 11:

Escribir un programa que reciba una cadena de texto y reporte:
a. Cuántas letras vocales en mayúscula se entraron 
b. Cuántas letras con tildes se entraron (minúsculas y mayúsculas)
c. Cuántos dígitos se entraron 
d. Cuántos espacios se entraron
e. Cuántas palabras reservadas se entraron
Utilizando para ello los métodos de cadenas especializados como .isupper(),
.isdigit(), etc. y la palabra reservada in.
"""
#%%
"""
Con la librería 'keywor' podemos tener las palabras dentro del código de manera
discriminada. Creo una variable para cada item que me piden, que me ayude a 
contar la cantidad de dígitos con las características y con el comando 'for' 
identificarlas. Luego siguiendo con el comando 'for', creo una condición que 
sume una unidad a las variables cada vez que se encuentre con un dígito con las
características señaladas en cada uno de los items.
"""

import keyword  
texto=input('Ingrese una cadena de texto: ')

contadorvocales= 0
contadortilde= 0
contadordigitos= 0
vocales_mayuscula= 'AEIOU'
letrascontilde= 'ÁáÉéÍíÓóÚú'
contadorespacio= 0
contador_reservadas= 0
lista=keyword.kwlist
listapalabras= texto.split()


for i in texto:
    if i in vocales_mayuscula:
        contadorvocales += 1
print('El texto tiene',contadorvocales,'vocales en mayúscula')
        
for i in texto:
    if i in letrascontilde:
        contadortilde += 1
print('El texto tiene',contadortilde,'letras con tilde')

for i in texto:
    if i.isdigit():
        contadordigitos +=1
print('El texto tiene',contadordigitos,'digitos')
       
for i in texto:
    if i in ' ':
        contadorespacio += 1
print('El texto tiene',contadorespacio,'espacios')

for i in listapalabras:
    if i in lista:
        contador_reservadas += 1
print('El texto tiene',contador_reservadas,'palabras reservadas' )
```
